```python
# Cell 2
import pandas as pd
import matplotlib.pyplot as plt
#
import warnings
warnings.filterwarnings('ignore')
warnings.simplefilter('ignore')
#
data_set = pd.read_csv("./train.csv")
print(data_set.head())
data_set.shape
data_set.describe()
X=data_set["GrLivArea"].dropna()
Y=data_set["SalePrice"].dropna()
plt.scatter(X,Y)
plt.xlabel("GrLivArea")
plt.ylabel("SalePrice")
plt.show()
```

## Cell 3

Review the scatter plot of GrLivArea versus SalePrice. Does there appear to be any outlier(s)? If so, please identify the point or points that you believe to be outliers. Do not remove the outliers from the dataframe.

**There are two points that appear to be outliers. The first has a value of GrLivArea of approximately 4500 and the second outlier has a GrLivArea value of approximately 5500.**

```python
# Cell 4
# enter the python code to fit a simple linear regression model using statsmodel

import statsmodels.formula.api as sm
print("Simple Linear Regression using Statmodels")

model1 = sm.ols(formula='SalePrice~GrLivArea',data=data_set)
fitted1 = model1.fit()
print(fitted1.summary())
```

##### Cell 5

Examine the results of the Python output from cell four and comment if the y-intercept and slope terms are statistically significant. You may use the value of alpha = 0.05 to make your decision.

**Both the y-intercept and slope terms have p-values of zero and are both statistically significant.**


```python
# Cell 6
# enter the Python code to construct and plot standardized residuals
import numpy as np

residual=fitted1.resid
residual2=fitted1.resid_pearson

plt.plot(data_set["GrLivArea"], residual2, 'ro')
plt.xlabel('GrLivArea')
plt.ylabel('standardized residual')
plt.title('Standardized Residuals versus GrLivArea')
plt.show()

fittedValues=fitted1.fittedvalues
plt.plot(fittedValues, residual2, 'ro')
plt.xlabel('fitted values')
plt.ylabel('standardized residual')
plt.title('Standardized Residuals versus fitted values')
plt.show()

id=np.arange(1,data_set["GrLivArea"].size+1,1)
plt.plot(id,residual2,'-bo')
plt.xlabel('index')
plt.ylabel('standardized residual')
plt.title('Standardized Residuals versus index')
plt.show()
```

```python
# Cell 10
# enter the Python code to perform an analysis of Cook's Distance

from statsmodels.stats.outliers_influence import OLSInfluence
test_class = OLSInfluence(fitted1)

plt.stem(np.arange(len(test_class.cooks_distance[0])),test_class.cooks_distance[0])
plt.show()
```
