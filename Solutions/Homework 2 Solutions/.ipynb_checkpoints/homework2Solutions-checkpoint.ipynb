```python
# Cell 2 - Answer Part A here
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Specify the path to your Excel file
file_path = 'wdbc_data.xlsx'

# Read the Excel file into a Pandas DataFrame
df = pd.read_excel(file_path)

# Display the first few rows of the DataFrame
#df.head()

#scatter matrix

pd.plotting.scatter_matrix(df,diagonal='kde')

# descriptive Statistics

df.describe()

# correlation analysis

# Calculate the correlation matrix
correlation_matrix = df.corr()
print("correlation matrix")
print(correlation_matrix)

# Display the correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title("Correlation Matrix")
plt.show()
```

## Cell 3 - Part B: Multicollinearity Discussion

Yes, multicollinearity will be a problem. It is easiest to observe in the Heat Map. All red cells indicate very strong positive correlation. The red-orange cells also indicate strong positive correlation. 

```python
# Cell 4 - Answer Part C here

from sklearn import preprocessing
std_scaler=preprocessing.StandardScaler().fit(df.values)
scaled_data=std_scaler.transform(df)
scaled_df=pd.DataFrame(scaled_data,index=df.index,columns=df.columns)
scaled_df.describe()
```

## Cell 5 - Part D Discussion

In reviewing the descriptive statistics for the scaled dataframe, all variables have zero mean and standard deviation of 1.0.

```python
# Cell 6 - Answer Part E here

import numpy as np
from sklearn import decomposition
#set up and execute PCA
pca=decomposition.PCA(n_components=10)
pca.fit(scaled_df)
#display eigen values
print("eigenvalues (sorted)")
print(pca.explained_variance_)
#display explained variance ratio
print("Percent variability explained by each component")
print(pca.explained_variance_ratio_)
#display cumulative sum of variability explained
print("Cumulative Percent variability explained by each component")
print(np.cumsum(pca.explained_variance_ratio_))
# display eigen vectors
print("Eigen Vectors")
print(pca.components_)
print("Scree Plot using Python")
plt.plot(np.arange(1,11,1),pca.explained_variance_,'-bo')
plt.show()
```

## Cell 7 - Part F Discussion

Using the eigen value criteria (>1), there would be two components.

From the Scree plot, I would recommend 4 components. This would account for 93.8% of the total variability.

```python
# Cell 8 - Answer Part G here

pca2=decomposition.PCA(n_components=4)
pca2.fit(scaled_df)
#display eigen values
print("eigenvalues (sorted)")
print(pca2.explained_variance_)
# display eigen vectors
print("Eigen Vectors")
print(pca2.components_)
```
