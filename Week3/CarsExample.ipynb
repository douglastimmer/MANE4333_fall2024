```python
# Cell One - Import Cars Dataset
import pandas as pd
cars = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data-original",
                  delim_whitespace = True, header=None,
                  names=['mpg','cylinders','displacement','horsepower','weight','acceleration','model','origin','car_name'])

cars.info()
```

```python
# cell two
# finding rows with missing values by column
#print(cars[cars['mpg'].isnull()])
#
cars.loc[cars.isnull().any(axis=1)]

```

```python
# cell three - graphical data screening
# first create data set with complete data
cars_complete=cars.dropna(axis=0)
cars_complete.info()
#
cars_complete.hist()
```

```python
# cell four boxplots

#cars_complete.boxplot()
cars_complete.boxplot(column=['horsepower'], return_type='axes')
```

```python
# cell five matrix of scatter plots
import pandas as pd
pd.plotting.scatter_matrix(cars_complete,diagonal='kde')
```

```python
# cell six - descriptive statistics

cars_complete.describe()
```

```python
# cell seven - MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
mpgScaler = MinMaxScaler()
mpgScaler.fit(cars_complete[['mpg']])
mpgMinMax=mpgScaler.transform(cars_complete[['mpg']])
type(mpgMinMax)
mpgMinMax_df=pd.DataFrame(mpgMinMax)
mpgMinMax_df.describe()
```

```python
# cell eight - StandardScaler
from sklearn.preprocessing import StandardScaler
weightScaler=StandardScaler()
weightScaler.fit(cars_complete[['weight']])
weightZscore = weightScaler.transform(cars_complete[['weight']])
weightZscore_df=pd.DataFrame(weightZscore)
weightZscore_df.columns = ['Z-score-weight']
weightZscore_df.describe()
```

```python
# cell nine - Decimal Scaling
from sklearn.preprocessing import MinMaxScaler
mpgDecimalScaler = MinMaxScaler(feature_range=(-1, 1))
mpgDecimalScaler.fit(cars_complete[['mpg']])
mpgDecimal=mpgDecimalScaler.transform(cars_complete[['mpg']])
mpgDecimal_df=pd.DataFrame(mpgDecimal)
mpgDecimal_df.columns=['mpg-Decimal']
mpgDecimal_df.describe()
```
